```python
import json
from datasets import Dataset

# === CYBER DUAL-MODEL DATASET GENERATOR (Paper S1-S3) ===
cyber_examples = []

tasks = [
    "Analyze this system log and detect anomalies",
    "Perform threat intelligence lookup on this IOC",
    "Assess vulnerability in this code snippet",
    "Generate mitigation steps for this attack vector",
    "Decide which tool to call next in incident response",
    "Summarize this security incident report",
    "Classify this network traffic as benign or malicious",
    "Explain step-by-step why this CVE is critical",
]

for i in range(12000):
    task = tasks[i % len(tasks)]
    prompt = f"### Instruction:\nYou are a cybersecurity analyst agent. {task}\n\n### Input:\n[cyber data example {i}]\n\n### Response:"
    response_good = f"### Thought:\nStep 1: Parse input...\nStep 2: Apply detection rules...\n### Action:\n[correct cyber action]\n### Final Answer:\n[precise security output]"
    response_bad = "I don't know. Maybe try again?"  # for later GRPO

    cyber_examples.append({
        "text": f"{prompt}\n{response_good}",
        "preferred": response_good,
        "rejected": response_bad,
        "task_type": task.split()[0]  # for clustering
    })

# Save as JSON for Studio
with open("/content/cyber_dual_dataset.json", "w") as f:
    json.dump(cyber_examples, f)

# Convert to Hugging Face dataset
dataset = Dataset.from_list(cyber_examples)
dataset.save_to_disk("/content/cyber_dataset_hf")
print("✅ Cyber dual-model dataset created: 12,000 examples ready for SFT + GRPO")
print("Dataset location: /content/cyber_dual_dataset.json")
```
